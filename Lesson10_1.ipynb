```python
import random
students = []
for _ in range(50):
    scores = []
    for _ in range(5):
        scores.append(random.randint(50, 100))
    students.append(scores)
students
```

```python
print("國文\t英文\t數學\t自然\t社會")
for score in students:
    for item in score:
        print(item, end="\t")
    print()
```
